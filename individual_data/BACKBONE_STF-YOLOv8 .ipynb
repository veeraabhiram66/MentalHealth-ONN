```python
from ultralytics import YOLO
config_path = '/home/ec2-user/anaconda3/envs/python3/lib/python3.10/ultralytics/cfg/models/v8/yolov8x_DW_swin_FOCUS-3.yaml'
data_path = '/home/ec2-user/SageMaker/Roboflow_DATA/data.yaml'
epochs = 100
img_size = 768
workers = 4
mosaic = 1
mixup = 0.5
flipud = 0.5
fliplr = 0.5
cache = True
save = True

# Instantiate YOLO model
model = YOLO('/home/ec2-user/SageMaker/runs/detect/train30/weights/best.pt')

# Train the model
results = model.train(resume= True,optimizer='AdamW',patience=2,data=data_path,save_period = 5, seed=0, batch=4, epochs=epochs, imgsz=img_size, workers=workers, mosaic=mosaic, mixup=mixup, flipud=flipud, fliplr=fliplr, cache=cache, save=save ) 
```
